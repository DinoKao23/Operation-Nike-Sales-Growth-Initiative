### Replace value in trasnaction file to PRODUCT_ID

We also need to merge the dataframe together with same key.
Therefore, we replace the ItemCode with Nike's product_id.

```python
# Check the data types
import random
```

```python
# Load the data with transaction data
df = pd.read_csv("transaction_data.csv")
df.head()
```

```python
# Check the transaction_data size
df.shape
```

```python
# Replace the ItemCode with the PRODUCT_ID


random.seed(27)
tst = old_df['PRODUCT_ID'].unique()
random_value= random.choices(tst, k = len(df))

df['ItemCode']= random_value
```

## Create Randomness in Dataframe to increase the Analysis Values

Original data only contains 3, 6, 9, 12, and 15 these five values.
With Randomness, we can change the range from 3-15 to 1-12.



```python
df["NumberOfItemsPurchased"] = df.apply(lambda row: random.randint(1, 12), axis=1)
```

## Replace User ID with Random Value
We create new customers from the Fake Customers generated from Other Source.
To make it more realistic, we replace the User ID with a random value.

```python
# Keep the original data's not logged in users
not_one = df.query('UserId != -1')
one = df.query('UserId == -1')
```

```python
# Original data only contains 4339 values
not_one['UserId'].nunique()
```

```python
# Replace the User ID with a random value between 100000 and 199999
not_one.loc[:,'UserId'] = df.apply(lambda row: random.randint(100000, 199999), axis=1)
```

```python
# merge the two dataframes togehter
result = pd.concat([not_one,one])
```

```python
# Merge the dataset together
result = pd.merge(result, old_df, how='left', left_on='ItemCode', right_on='PRODUCT_ID')
```

```python
# Remove redundant columns 
result.columns
```

```python
result.drop(['CURRENCY','TID', 'PRODUCT_URL', 'CostPerItem', 'PRODUCT_SIZE','InsertUpdateTime', 'RunDate', 'PRICE_CURRENT'], axis = 1, inplace= True)
```

```python
# Make a calculation field for the total revenue
result['revenue'] = result['PRICE_RETAIL'] * result['NumberOfItemsPurchased']
```
